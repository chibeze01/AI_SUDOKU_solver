## Part One
You should write all of your code for solving sudokus below this cell.

You must include a function called `sudoku_solver(sudoku)` which takes one sudoku puzzle (a 9x9 NumPy array) as input, and returns the solved sudoku as another 9x9 NumPy array. This is the function which will be tested. 

```python
import copy
def sudoku_solver(sudoku):
    """
    Solves a Sudoku puzzle and returns its unique solution.

    Input
        sudoku : 9x9 numpy array
            Empty cells are designated by 0.

    Output
        9x9 numpy array of integers
            It contains the solution, if there is one. If there is no solution, all array entries should be -1.
    """
    
    ### YOUR CODE HERE
    
    solved_sudoku = copy.copy(sudoku)
    # sudoku puzzel width
    
    length  = solved_sudoku[0,:].size
    
    n = np.array ([i for i in range(1,length+1)])


#     value_grid = copy.copy(sudoku[p]).tolist()

    while 0 in solved_sudoku:
        for i in range (0,length):
            r = solved_sudoku[i]
            
            # check if the puzzel has duplicates values in a row
            unique_arr, count_arr = np.unique(r[r!=0], return_counts=True)
            if unique_arr[count_arr>1].size > 0: return np.full((length, length), -1.)
#             print (r)

            for j in range (0,length):
                c = solved_sudoku[:,j]
                
                # check if the puzzel has duplicates values in a column
                unique_arr, count_arr = np.unique(c[c!=0], return_counts=True)
                if unique_arr[count_arr>1].size > 0 :return np.full((length, length), -1.)
                
                if solved_sudoku[i][j] == 0 :   
#                     print(c)

                    nc = np.setdiff1d(n,c)  
                    nr = np.setdiff1d(n,r)

                    possible_v = np.intersect1d(nc,nr)    

#                     print(possible_v, nr, nc)

#                     value_grid[i][j] = possible_v.tolist()
                    if possible_v.size == 1:
                        solved_sudoku[i][j] = possible_v[0]
                    if possible_v.size == 0:
                        return np.full((length, length), -1.)


#     print(value_grid)

    
    return solved_sudoku
your_solution = sudoku_solver(sudokus[1])
# print(sudokus[0], your_solution)
if np.array_equal(your_solution, solutions[1]):
    print("Yes! Correct solution.")
else:
    print("No, the correct solution is:")
    print(solutions[0])
```
